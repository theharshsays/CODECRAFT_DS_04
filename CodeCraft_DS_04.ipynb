```python
import matplotlib.pyplot as plt

# Example data for visualization
sentiments = {
    'Assassins Creed': {'Positive': 5, 'Negative': 0, 'Neutral': 0, 'Irrelevant': 0},
    'Call of Duty': {'Positive': 3, 'Negative': 4, 'Neutral': 2, 'Irrelevant': 2},
    'Dota 2': {'Positive': 1, 'Negative': 1, 'Neutral': 0, 'Irrelevant': 0},
    # Add more data here...
}

# Bar chart for sentiment distribution
plt.figure(figsize=(10, 6))
for i, (brand, sentiment_counts) in enumerate(sentiments.items()):
    plt.bar(range(4), list(sentiment_counts.values()), label=brand)
    plt.xticks(range(4), ['Positive', 'Negative', 'Neutral', 'Irrelevant'])
    plt.title('Sentiment Distribution')
    plt.legend()
plt.show()

# Pie chart for a specific brand
brand = 'Call of Duty'
plt.figure(figsize=(6, 6))
plt.pie(sentiments[brand].values(), labels=sentiments[brand].keys(), autopct='%1.1f%%')
plt.title(f'Sentiment Distribution for {brand}')
plt.show()
```
